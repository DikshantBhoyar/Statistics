**1. What is hypothesis testing in statistics ?**
 -  Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It's a way of testing a claim or assumption (called a hypothesis) about a population parameter.
---
**2.  What is the null hypothesis, and how does it differ from the alternative hypothesis ?**

  - Null hypothesis (H0): The null hypothesis is the default assumption. It says that there is no effect, no difference, or no change. It assumes everything is normal or as expected.
- It is what you're testing against.  
- It usually uses signs like equals, less than or equal to, or greater than or equal to.

- Example:  
If you're testing a medicine's effect on blood pressure:  
H0: The pill has no effect on blood pressure.


 - Alternative hypothesis (H1 or Ha): The alternative hypothesis is what you're trying to prove. It says there is an effect, a difference, or a change.
- It is the opposite of the null hypothesis.  
- It usually uses signs like not equal to, less than, or greater than.

- Example :  
H1: The pill does affect blood pressure.

---
**3.  What is the significance level in hypothesis testing, and why is it important ?**

  - The significance level in hypothesis testing is a value that helps you decide whether to reject the null hypothesis or not.
- It is usually written as alpha (α).  
- It represents the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.


 - Common values for α are:
- 0.05 (5%)  
- 0.01 (1%)  
- 0.10 (10%)


- It is important because it sets the threshold for how much evidence you need from your sample to reject the null hypothesis.


 - For example:
- If α = 0.05, you're saying you're willing to accept a 5% chance of being wrong when you reject the null hypothesis.  
- If your p-value (from the test) is less than α, you reject the null hypothesis.


 - Example:
- Suppose you're testing if a new drug works better than the old one.  
You set α = 0.05.  
After testing, your p-value is 0.03.
- Since 0.03 < 0.05, you reject the null hypothesis and say the new drug is significantly better.

- The significance level is the line you draw to decide whether the result you got is just by chance or actually meaningful.

---
**4. What does a P-value represent in hypothesis testing ?**
   - The p-value in hypothesis testing represents the probability of getting results as extreme as (or more extreme than) the ones you observed, assuming that the null hypothesis is true.It tells us how likely our sample results are just due to chance.


 - Interpretation :
- If the p-value is small (like less than 0.05), it means your results are unlikely to have happened by chance, so you reject the null hypothesis.  
- If the p-value is large (like more than 0.05), it means your results could have easily happened by chance, so you do not reject the null hypothesis.

- Example:

Let’s say we are testing whether a coin is fair.

Null hypothesis: The coin is fair.  
We flip it 100 times and get 70 heads.  
We calculate the p-value and it turns out to be 0.01.

Since 0.01 is less than 0.05, the result is unlikely to happen by chance, so we reject the null hypothesis and say the coin is probably biased.

So, the p-value helps you measure the strength of evidence against the null hypothesis.


---

**5. How do you interpret the P-value in hypothesis testing ?**
  - The p-value tells you how likely it is to get the observed results, or something more extreme, if the null hypothesis is true.
- If the p-value is small, it means the observed results are unlikely under the assumption that the null hypothesis is true.  
- If the p-value is large, it means the observed results are quite likely under the null hypothesis.


 - Interpretation :
- If the p-value is less than or equal to your chosen significance level (for example, 0.05), you reject the null hypothesis. This means your result is statistically significant.
- If the p-value is greater than your significance level, you do not reject the null hypothesis. This means there is not enough evidence to say the result is significant.

- Example:

You test a new medicine and get a p-value of 0.03.  
If your significance level is 0.05, then 0.03 is less than 0.05.  
So, you reject the null hypothesis and conclude that the medicine has a significant effect.

In short:

A smaller p-value means stronger evidence against the null hypothesis.  
A larger p-value means weaker evidence against the null hypothesis.

**6. What are Type 1 and Type 2 errors in hypothesis testing ?**
   - Type 1 and Type 2 errors are two kinds of mistakes that can happen when making decisions in hypothesis testing.


 - Type 1 Error :
- This happens when you reject the null hypothesis even though it is actually true.  
- It means you think there is an effect or difference when there really isn’t one.


 - Example:  
- You test a new drug and conclude it works better than the old one, but in reality, there is no difference.  
- You made a Type 1 error.

  The probability of making a Type 1 error is called alpha (significance level), usually set at 0.05.


 - Type 2 Error :
- This happens when you fail to reject the null hypothesis even though it is actually false.  
- It means you think there is no effect or difference when there actually is one.


 - Example :  
- You test a new drug and conclude it is not better than the old one, but in reality, it is.  
- You made a Type 2 error.

The probability of making a Type 2 error is called beta.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?**
   - One-tailed test:
  - A one-tailed test is used when you want to check if a value is either greater than or less than a certain number, but not both.
  - You are only interested in one direction of change.

- Example:  
You want to test if a new drug increases recovery speed compared to the old one.  
You are only checking if it is better, not worse.  
So, you use a one-tailed test.



- Two-tailed test :
 - A two-tailed test is used when you want to check if a value is simply different from a certain number, without caring whether it is higher or lower.
 - You are interested in both directions of change.

- Example:  
You want to test if a new drug has a different effect compared to the old one.  
It could be better or worse — you just want to know if it’s different.  
So, you use a two-tailed test.

**8. What is the Z-test, and when is it used in hypothesis testing ?**
  - The Z-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between sample data and a population parameter, or between two samples.


 - When is the Z-test used :
- The sample size is large (usually n > 30).  
- The population standard deviation is known.  
- The data is approximately normally distributed or the sample is large enough for the Central Limit Theorem to apply.

 - Common uses of the Z-test :
- To test whether the sample mean is different from a known population mean.  
- To compare the means of two large samples.  
- To test proportions.

- Example :

A factory claims that the average weight of a product is 100 grams.  
You take a sample of 50 items, and the average weight is 98 grams.  
You know the population standard deviation is 4 grams.  
You can use a Z-test to check if this difference is statistically significant.


- The Z-test helps you decide if the difference you see is real or just due to chance, when you know the population standard deviation and have a large enough sample.

**9.  How do you calculate the Z-score, and what does it represent in hypothesis testing ?**
 - How to calculate the Z-score :

The formula for the Z-score is:

Z = (X̄ − μ) / (σ / √n)

Where:  
- X̄ is the sample mean  
- μ is the population mean  
- σ is the population standard deviation  
- n is the sample size



 - Z-score represents :
- The Z-score tells you how many standard deviations your sample mean is away from the population mean, assuming the null hypothesis is true.

- In hypothesis testing, the Z-score helps determine whether the observed result is unusual or expected under the null hypothesis.

- Example :

Let’s say:  
- Population mean (μ) = 100  
- Population standard deviation (σ) = 10  
- Sample mean (X̄) = 104  
- Sample size (n) = 25

Z = (104 − 100) / (10 / √25)  
Z = 4 / (10 / 5) = 4 / 2 = 2

So, the Z-score is 2. This means the sample mean is 2 standard deviations above the population mean.


- How to use it :

Once you get the Z-score, you compare it to the critical Z-value (based on your chosen significance level, like 0.05) to decide whether to reject the null hypothesis.

**10. What is the T-distribution, and when should it be used instead of the normal distribution ?**
  - The t-distribution is a type of probability distribution that is very similar to the normal distribution but has heavier tails. This means it gives more probability to values far from the mean, which helps when dealing with smaller samples.

- The t-distribution be used :


 - You should use the t-distribution instead of the normal distribution when:
- The sample size is small (usually less than 30)  
- The population standard deviation is unknown  
- The data is approximately normally distributed or close to normal

 - Why use it?
- When the sample is small and you don’t know the population standard deviation, the estimates of variability are less reliable. The t-distribution accounts for this extra uncertainty by being wider (with thicker tails) than the normal distribution.

- Example :

Let’s say you are testing the average height of students in a class of 15.  
You don’t know the population standard deviation.  
Because your sample is small and the standard deviation is unknown, you should use the t-distribution.

- Use the normal distribution when the sample is large and the population standard deviation is known.  
- Use the t-distribution when the sample is small and the population standard deviation is unknown.

  - Z-test:

- Based on the standard normal distribution

  - T-test:
- Used when the sample size is small (typically 30 or less)

- Based on the t-distribution, which has heavier tails to account for more uncertainty

  - Key difference:
- The main difference is whether you know the population standard deviation and the size of your sample.  
- If you know the population standard deviation and have a large sample, use the Z-test.  
- If you don’t know the population standard deviation and have a small sample, use the T-test.

  - Example:
- You want to test the average weight of a product.

**12.  What is the T-test, and how is it used in hypothesis testing?**
 - The t-test is a statistical method used in hypothesis testing to determine whether there is a significant difference between the means of two groups or between a sample mean and a known value.

- It is especially useful when the sample size is small and the population standard deviation is unknown.


 - How the t-test is used in hypothesis testing:
- Set up the null hypothesis (usually assumes no difference) and the alternative hypothesis (assumes a difference exists).
- Choose a significance level (commonly 0.05).
- Calculate the t-statistic using the sample data.
- Find the critical t-value from the t-distribution table based on the sample size (degrees of freedom).

- Compare the t-statistic with the critical t-value.
   - If the t-statistic is more extreme than the critical value, reject the null hypothesis.
   - If not, do not reject the null hypothesis.

  
  - Types of t-tests:
- One-sample t-test: Compares the sample mean to a known population mean.
- Two-sample (independent) t-test: Compares the means of two independent groups.
- Paired t-test: Compares means from the same group at different times (like before and after treatment).


- Example:
 - Suppose a teacher wants to know if a new teaching method is more effective.  
She tests the scores of 15 students before and after using the new method.  
She uses a paired t-test to compare the average scores and see if the difference is statistically significant.

**13. What is the relationship between Z-test and T-test in hypothesis testing ?**
- The Z-test and T-test are both statistical methods used in hypothesis testing to determine if there's a significant difference between means or proportions, but they are used under different conditions.


 - Purpose is similar:  
- Both the Z-test and T-test are used to test hypotheses about population means and sometimes proportions.  
- They both help you decide whether to reject the null hypothesis.


 - Main difference lies in sample size and standard deviation:
     - Use the Z-test when:  
  - The sample size is large (typically more than 30)  
  - The population standard deviation is known
     - Use the T-test when:  
  - The sample size is small (typically 30 or less)  
  - The population standard deviation is unknown

 - Both rely on normal distribution:  
- The Z-test uses the standard normal distribution.  
- The T-test uses the t-distribution, which is similar to the normal distribution but has heavier tails to account for more variability with smaller samples.

 - As sample size increases:  
- The t-distribution becomes very close to the normal distribution.  
So, for large samples, both tests give nearly the same result, and you can use either.

**14. What is a confidence interval, and how is it used to interpret statistical results ?**
  - A confidence interval is a range of values, calculated from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.

- How it is interpreted :
 - If you say you have a 95 percent confidence interval of [45, 55] for the average score of students, it means you are 95 percent confident that the true average score of all students lies somewhere between 45 and 55.
   - Note: It does not mean there is a 95 percent chance that the true mean is between 45 and 55. Instead, it means that if you repeated the experiment many times, 95 percent of the calculated intervals would contain the true mean.


 - How it's used in statistical results:
- Estimation:  
     - Instead of giving just a single estimate (like a sample mean), a confidence interval gives a range, which is more informative.
- Decision making:  
     - If a confidence interval does not contain a certain value (like 0 or a target mean), it might suggest a statistically significant difference.

- Example:  
 - If a 95 percent confidence interval for the difference in means is [2.1, 5.3], and it does not include 0, you can conclude there’s a significant difference between the groups.

- Confidence level (like 90 percent, 95 percent, 99 percent) tells you how confident you are in the method. Higher levels give wider intervals, but more confidence.

**15.  What is the margin of error, and how does it affect the confidence interval ?**
 -  The margin of error is the amount added and subtracted from a sample estimate (like a sample mean or proportion) to create a confidence interval. It reflects the maximum expected difference between the true population value and the sample estimate due to sampling variability.

 - It affects the confidence interval:
- The margin of error determines the width of the confidence interval. A larger margin of error means a wider interval, giving a less precise estimate. A smaller margin of error gives a narrower, more precise interval.

- Example:  
If the sample mean is 100 and the margin of error is 5,  
then the 95 percent confidence interval is:  
100 ± 5 → [95, 105]


 - Influences the margin of error:
- Sample size : Larger samples reduce the margin of error.  
- Confidence level : Higher confidence levels (like 99 percent vs 95 percent) increase the margin of error.  
- Data variability : More variability in the data leads to a larger margin of error.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**
- Bayes' Theorem is a fundamental concept in statistics used to update the probability of a hypothesis based on new evidence or information. It connects prior knowledge with new data to make better-informed decisions or predictions.


 - It is used in statistics:
- Bayes' Theorem helps calculate the posterior probability, which is the probability of an event or hypothesis given the observed data.

The formula is:

P(A | B) = [P(B | A) × P(A)] / P(B)

Where:  
P(A | B) is the probability of event A given that B has occurred (posterior probability)  
P(B | A) is the probability of event B given that A has occurred (likelihood)  
P(A) is the initial probability of A being true (prior probability)  
P(B) is the overall probability of B (marginal likelihood)


 - Example:
- Suppose 1 percent of people have a disease. A test detects the disease 99 percent of the time if someone has it, but also falsely detects it in 5 percent of healthy people.
    - If a person tests positive, Bayes' Theorem helps calculate the actual chance they have the disease, which will likely be much lower than 99 percent, due to the low base rate (1 percent).


 - Significance:
- It improves decision-making by combining past information with new data.  
- It is used in medical testing, spam filtering, machine learning, finance, and more.  
- It supports a more probabilistic way of reasoning, especially when dealing with uncertainty.

**17. What is the Chi-square distribution, and when is it used?**
  - The Chi-square distribution is a type of probability distribution that is used in statistics to analyze categorical data. It is commonly used in tests that involve counts or frequencies, not means.

- The Chi-square distribution is a distribution of the sum of squared standard normal variables. It is always positive and skewed to the right, especially with fewer degrees of freedom. As the degrees of freedom increase, it becomes more symmetric.


 - It is used:
- Chi-square test of independence:  
       - Used to determine whether there is a significant association between two categorical variables.  
Example: Checking if gender and product preference are related.
- Chi-square goodness-of-fit test:  
      - Used to see if observed frequencies match expected frequencies.  
Example: Testing if a die is fair by comparing the observed roll outcomes to what you'd expect.
- Test of variance:  
      - Sometimes used to compare the variance of a sample to a known value when the data is normally distributed.


- It deals with categorical data such as colors, brands, or categories  
- The test statistic measures how much the observed counts differ from the expected counts  
- A large Chi-square value indicates that the observed and expected frequencies are different enough to suggest a statistically significant difference

**18.  What is the Chi-square goodness of fit test, and how is it applied?**
  - The Chi-square goodness of fit test is a statistical test used to determine whether the observed frequencies of a categorical variable match the expected frequencies based on a specific hypothesis.


 - Purpose:
- It checks if a sample data distribution fits a theoretical or expected distribution.


 - How it is applied:
- State the hypotheses:  
    - Null hypothesis (H0): The observed frequencies match the expected frequencies.  
    - Alternative hypothesis (H1): The observed frequencies do not match the expected frequencies.
- Calculate the expected frequencies:  
    - These are usually based on a known or assumed distribution, like equal probability for each category.
- Use the Chi-square formula:  

Chi-square = Σ [(O - E)² / E]  

Where:  
O is the observed frequency  
E is the expected frequency  
Σ means to sum across all categories

  - Determine the degrees of freedom:  
- Degrees of freedom = number of categories minus 1

  - Find the critical value or p-value:  
- Compare the calculated Chi-square value with the critical value from the Chi-square distribution table based on your chosen significance level, like 0.05.

  - Make a decision:  
- If the Chi-square statistic is greater than the critical value, reject the null hypothesis.  
Otherwise, do not reject it.

Example:

Suppose a die is rolled 60 times and the results are:

Face 1: 8 times  
Face 2: 10 times  
Face 3: 9 times  
Face 4: 12 times  
Face 5: 11 times  
Face 6: 10 times  

Expected frequency for each face (if fair) is 60 divided by 6, which equals 10

Use the formula to calculate the Chi-square value and see if the differences are due to chance or if they suggest the die is biased.

**19. What is the F-distribution, and when is it used in hypothesis testing?**
  - The F-distribution is a probability distribution that arises frequently in statistics when comparing variances between two populations. It is used primarily in hypothesis testing for analyzing whether group variances are significantly different.

- It is right-skewed and only takes positive values.  
- It is defined by two different degrees of freedom: one for the numerator and one for the denominator.  
- The shape of the distribution depends on these degrees of freedom.


 - It is used in hypothesis testing:
- ANOVA (Analysis of Variance):  
    - The most common use of the F-distribution is in ANOVA tests, where it compares the variance between group means to the variance within the groups.  
Example: Testing if three different diets lead to different average weight loss.
- Comparing Two Variances:  
     - An F-test can compare whether two population variances are significantly different from each other.
- Regression Analysis:  
     - In multiple linear regression, the F-test checks if the overall regression model is significant, meaning whether at least one predictor variable has a meaningful effect.

- The F-statistic is calculated as:  
F = Variance between groups divided by Variance within groups

A larger F value suggests a greater difference between groups, which might be statistically significant.  
If the F value is close to 1, it suggests the group variances are similar, supporting the null hypothesis.

**20. What is an ANOVA test, and what are its assumptions?**
  - An ANOVA test (Analysis of Variance) is a statistical method used to determine whether there are significant differences between the means of three or more independent groups.

- ANOVA helps test the null hypothesis that all group means are equal, against the alternative hypothesis that at least one group mean is different.


 - Example:  
- You want to compare the average scores of students from three different teaching methods to see if any method leads to better performance.


 - Types of ANOVA:
- One-way ANOVA – Compares means of groups based on one independent variable.  
- Two-way ANOVA – Examines the effect of two independent variables and their interaction.  
- Repeated measures ANOVA – Used when the same subjects are tested more than once.

- ANOVA Assumptions:
 - Independence of observations:  
   - The data collected from different groups should be independent.
 - Normality:  
   - The data in each group should be approximately normally distributed.
 - Homogeneity of variances (homoscedasticity):  
   - The variances of the groups should be roughly equal.
 - It calculates the F-statistic, which is the ratio of variance between groups to variance within groups.  
 - A high F value suggests significant differences between at least some group means.  
 - If the result is significant, a post hoc test (like Tukey’s test) is usually performed to find out which specific groups differ.
---
**21. What are the different types of ANOVA tests?**
- One-Way ANOVA :
   - Purpose: Used to compare the means of three or more independent groups based on one independent variable (factor).
   - Example: Comparing the average test scores of students from three different teaching methods.
- Two-Way ANOVA:
   - Purpose: Used when there are two independent variables. It examines the individual and combined effects of both variables on the dependent variable.
   - Example: Studying how different teaching methods (first independent variable) and student study time (second independent variable) affect test scores.
   - It also helps to assess if there is an interaction effect between the two independent variables.
- Repeated Measures ANOVA:
   - Purpose: Used when the same subjects are tested multiple times (i.e., repeated measurements from the same group). It accounts for the correlation between repeated measurements.
   - Example: Measuring the test scores of students before, during, and after a training program.
   - This type of ANOVA helps to see if there are differences across the different time points or conditions.

**22. What is the F-test, and how does it relate to hypothesis testing?**
  - The F-test is a statistical test used to compare variances between two or more groups and determine if the variances are significantly different. It is most commonly used in the context of ANOVA (Analysis of Variance) but can also be used for other purposes, such as comparing the goodness of fit of regression models.


 - Purpose of the F-test:
- It assesses whether the variances of two or more populations are equal.
- It is used to test the hypothesis that the means of multiple groups are equal, which is an essential part of the ANOVA.
- The F-test can also compare the fit of different statistical models, such as comparing a full model and a reduced model in regression analysis.



 - The F-test works:
- Null hypothesis (H₀):  
    - The variances (or group means) are equal across the groups. In the context of ANOVA, this means there is no significant difference in the means of the groups being tested.
- Alternative hypothesis (H₁):  
    - At least one group mean is different, or the variances are not equal.
- F-statistic formula:  
    - F = Variance between groups / Variance within groups

    - Variance between groups reflects the variability due to the treatment or factor being tested (for example, the different groups or conditions).
    - Variance within groups reflects the variability within each group due to random error or individual differences.
-  Degrees of freedom (df):  
    - Numerator df: Related to the number of groups or treatments (k - 1).
    - Denominator df: Related to the number of observations in each group (N - k).
- Decision rule:  
     - Once the F-statistic is calculated, it is compared to a critical value from the F-distribution table based on the degrees of freedom and the chosen significance level (alpha, e.g., 0.05).
     - If the calculated F-statistic is greater than the critical value, the null hypothesis is rejected, suggesting that there is a significant difference between the groups or that the model is a better fit.

```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std):
    sample_mean = np.mean(sample_data)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample_data)))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    return z_stat, p_value

sample_data = [15, 16, 17, 14, 19, 16, 17, 18, 15, 16]   # Example data
population_mean = 16
population_std = 2

z_stat, p_value = z_test(sample_data, population_mean, population_std)

alpha = 0.05      # Result interpretation
print(f"Z-statistic: {z_stat}, P-value: {p_value}")
print("Reject null hypothesis" if p_value < alpha else "Fail to reject null hypothesis")
```

```python
#Q3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample_data, population_mean, population_std):    # Function to perform the one-sample Z-test
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))     # Calculate Z-statistic

    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))    # Calculate p-value (two-tailed test)

    return z_stat, p_value

# Example sample data (replace with your data)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample of size 30 with mean 52

population_mean = 50  # Known population parameters
population_std = 10

z_stat, p_value = one_sample_z_test(sample_data, population_mean, population_std)  # Perform the one-sample Z-test

alpha = 0.05  # Significance level
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

if p_value < alpha:   # Interpretation of the results
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_errors(pop_mean, std, size, alpha=0.05, true_mean=None):
    crit_value = stats.norm.ppf(1 - alpha / 2)

    z_null = (np.mean(np.random.normal(pop_mean, std, size)) - pop_mean) / (std / np.sqrt(size))        # Type 1 Error (Null hypothesis true)
    type_1 = abs(z_null) > crit_value

    type_2 = False          # Type 2 Error (Alternative hypothesis true)
    if true_mean:
        z_alt = (np.mean(np.random.normal(true_mean, std, size)) - pop_mean) / (std / np.sqrt(size))
        type_2 = abs(z_alt) <= crit_value

    # Plot
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, stats.norm.pdf(x))
    plt.fill_between(x, stats.norm.pdf(x), where=(x < -crit_value) | (x > crit_value), color='red', alpha=0.5)
    plt.axvline(crit_value, color='black', linestyle='--')
    if type_1: plt.axvline(z_null, color='green', linestyle='--')
    if type_2: plt.axvline(z_alt, color='orange', linestyle='--')
    plt.title("Type 1 & Type 2 Errors")
    plt.show()

    return type_1, type_2

# Example usage
pop_mean = 50
std = 10
size = 30
true_mean = 55
type_1, type_2 = calculate_errors(pop_mean, std, size, true_mean=true_mean)

print(f"Type 1 Error: {type_1}, Type 2 Error: {type_2}")
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error
    margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err
    return mean - margin, mean + margin

np.random.seed(42)
sample_data = np.random.normal(100, 15, 40)

# Calculate 95% confidence interval
ci_low, ci_high = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

```python
#Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)  # Standard error
    t_score = stats.t.ppf((1 + confidence) / 2, df=n - 1)
    return t_score * std_err

np.random.seed(42)
sample_data = np.random.normal(100, 12, 50)

# Calculate margin of error for 95% confidence level
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, sensitivity, specificity):
    # P(D|H): Likelihood (True positive rate)
    # P(D|~H): False positive rate = 1 - specificity
    p_positive = sensitivity * prior + (1 - specificity) * (1 - prior)
    posterior = (sensitivity * prior) / p_positive
    return posterior

# Given probabilities
prior = 0.01         # P(H) - 1% have the disease
sensitivity = 0.99   # P(Positive | Has disease)
specificity = 0.95   # P(Negative | No disease)

posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of having the disease given a positive test: {posterior:.2%}")
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

np.random.seed(42)      # Generate two random samples with different variances
sample1 = np.random.normal(0, 2, 100)  # Mean 0, Std dev 2
sample2 = np.random.normal(0, 4, 100)  # Mean 0, Std dev 4

# Calculate the sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F_stat = max(var1, var2) / min(var1, var2)     # F-statistic: ratio of the larger variance to the smaller variance

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

p_value = 2 * min(f.cdf(F_stat, df1, df2), 1 - f.cdf(F_stat, df1, df2))   # Calculate the p-value for the F-test

print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#Q19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    """
    Checks normality and homogeneity of variances for given groups.
    Independence is assumed if samples are random and independent.
    """
    print("---- Normality (Shapiro-Wilk Test) ----")
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i}: W={stat:.4f}, p={p:.4f} ->",
              "Normal" if p > 0.05 else "Not normal")

    print("\n---- Homogeneity of Variance (Levene’s Test) ----")
    stat, p = levene(*groups)
    print(f"Levene’s Statistic={stat:.4f}, p={p:.4f} ->",
          "Equal variance" if p > 0.05 else "Unequal variance")

    print("\n---- Independence ----")
    print("Assumed if data collected independently (not testable here).")

    # Optional: Visual normality check
    plt.figure(figsize=(10, 4))
    for i, group in enumerate(groups, 1):
        plt.subplot(1, len(groups), i)
        plt.hist(group, bins=10, edgecolor='black')
        plt.title(f'Group {i}')
        plt.xlabel('Value')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.suptitle("Histograms for Normality Check", y=1.05)
    plt.show()


# Simulate three groups
group1 = np.random.normal(100, 10, 30)
group2 = np.random.normal(105, 10, 30)
group3 = np.random.normal(110, 10, 30)

check_anova_assumptions(group1, group2, group3)
```

```python
#Q24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

# Sample data
np.random.seed(42)
data = np.random.normal(loc=100, scale=12, size=30)  # std=12 → variance = 144
sample_size = len(data)
sample_var = np.var(data, ddof=1)  # unbiased sample variance

# Hypothesized population variance
sigma_0_squared = 144  # assumed population variance

# Chi-square test statistic
chi2_stat = (sample_size - 1) * sample_var / sigma_0_squared

# Degrees of freedom
df = sample_size - 1

# Critical values for 95% confidence level (two-tailed)
alpha = 0.05
lower_crit = chi2.ppf(alpha / 2, df)
upper_crit = chi2.ppf(1 - alpha / 2, df)

# P-value
p_val = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

# Results
print(f"Sample variance: {sample_var:.2f}")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"Critical values: ({lower_crit:.2f}, {upper_crit:.2f})")
print(f"P-value: {p_val:.4f}")

# Interpretation
if chi2_stat < lower_crit or chi2_stat > upper_crit:
    print(" Reject the null hypothesis: Variance is significantly different.")
else:
    print(" Fail to reject the null hypothesis: No significant difference in variance.")
```

```python
#Q25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Example data:
# Group A: 50 successes out of 200
# Group B: 70 successes out of 220

successes = np.array([50, 70])
samples = np.array([200, 220])

# Perform two-proportion Z-test (two-tailed by default)
z_stat, p_val = proportions_ztest(successes, samples)

# Output results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Proportions are significantly different (reject H0).")
else:
    print(" No significant difference in proportions (fail to reject H0).")
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate data
np.random.seed(0)
a = np.random.normal(0, 10, 30)
b = np.random.normal(0, 15, 30)

# Variances & F-statistic
var_a, var_b = np.var(a, ddof=1), np.var(b, ddof=1)
f_stat = max(var_a, var_b) / min(var_a, var_b)
dfn, dfd = 29, 29
p_val = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

# Print
print(f"F = {f_stat:.2f}, p = {p_val:.4f}")
print(" Variances differ." if p_val < 0.05 else " No variance difference.")

# Plot
x = np.linspace(0, 5, 500)
plt.plot(x, f.pdf(x, dfn, dfd), label="F-distribution")
plt.axvline(f_stat, color='red', label='F-stat')
plt.title("F-Test: Variance Comparison")
plt.legend(), plt.grid(), plt.show()
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare

# Simulated observed data (e.g. outcomes of rolling a die 60 times)
observed = np.array([8, 10, 9, 12, 11, 10])
expected = np.full(6, 10)  # expected frequency = 60 rolls / 6 sides = 10

# Chi-square test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

# Results
print(f"Chi2 = {chi2_stat:.2f}, p = {p_val:.4f}")
if p_val < 0.05:
    print(" Reject H₀: Observed frequencies significantly differ from expected.")
else:
    print(" Fail to reject H₀: No significant difference from expected.")
```
